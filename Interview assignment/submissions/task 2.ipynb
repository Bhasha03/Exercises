```python
def calculation_distance_matrix(data):
    m = data.shape[0]
    output = pd.DataFrame(columns=data.iloc[:,0], index=data.iloc[:,0])
    output.columns.name = None
    output.index.name = None
    for i in range(m):
        output.iloc[i,i] = 0
        for j in range(i+1, m):
            count = np.sum(data.iloc[:,2][i:j])
            output.iloc[i,j] = count
            output.iloc[j,i] = count
    return output
```
