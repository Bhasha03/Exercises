```python


# Create the inventory dataset
inventory_df = pd.DataFrame({
    "Item Number": [1, 2, 3],
    "Current Inventory": [10, 20, 30]
})

# Create the new inventory dataset
new_inventory_df = pd.DataFrame({
    "Item Number": [1, 2, 3],
    "Arrival Date": ["2023-11-23", "2023-11-24", "2023-11-25"],
    "New Inventory": [5, 10, 15]
})

# Merge the two datasets
merged_df = inventory_df.merge(new_inventory_df, on="Item Number")

# Create a dictionary to store the inventory levels at different dates in the future
future_inventory_dict = {}

# Iterate through each item
for item_number in merged_df["Item Number"].unique():
    # Get the current inventory for the item
    current_inventory = merged_df[merged_df["Item Number"] == item_number]["Current Inventory"].values[0]

    # Create a list to store the inventory levels at different dates in the future
    future_inventory_list = []

    # Iterate through each row in the merged dataset for the current item
    for index, row in merged_df[merged_df["Item Number"] == item_number].iterrows():
        # Get the arrival date and new inventory for the row
        arrival_date = row["Arrival Date"]
        new_inventory = row["New Inventory"]

        # Calculate the inventory level at the arrival date
        inventory_level = current_inventory + new_inventory

        # Add the inventory level and arrival date to the list
        future_inventory_list.append((arrival_date, inventory_level))

    # Update the dictionary with the inventory levels for the current item
    future_inventory_dict[item_number] = future_inventory_list

# Create a pandas DataFrame from the dictionary
future_inventory_df = pd.DataFrame(future_inventory_dict)

# Print the DataFrame
print(future_inventory_df)
```
